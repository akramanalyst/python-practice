# How to import it:


```python
import matplotlib.pyplot as plt
```

# Line Plot

```python
x = [1,2,3,4]
y = [10,20,25,30]
plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(color='black')
plt.show()
```

# Adding Label  Titles and Grid:-

```python
x = [1,2,3,4]
y = [10,20,50,30]
plt.plot(x,y)
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.title("Line Chart")
plt.grid(True)
plt.show()
```

# styling the Line :-

```python
plt.plot(x,y, color = "red",linestyle="--",marker = "o")
plt.show()
```

# Common Options:

```python
---> color = "blue","red","green"
---> linestyle = "-","--",":"
---> marker = "o","s","^"
```

```python
x = [1,2,3,4,5]
y = [10,15,40,60,45]
plt.plot(x,y,color = "blue",linestyle = ":",marker = "s")
plt.show()
```

# Multiple Lines on one Plot:-

```python
y2 = [5,15,20,35,55]
plt.plot(x,y,label = "Line A",linestyle = "--",marker = "^")
plt.plot(x,y2,label = "Line B",linestyle = ":",marker = "o")
plt.legend()
plt.grid("True")
plt.show()
```

# Bar Chart:


```python
names = ["A","B","C"]
values = [10,24,36]
plt.bar(names,values,color = "orange")
plt.title("Bar Chart Example")
plt.xlabel("Names")
plt.ylabel("Values")
plt.show()
```

# Scatter Plot :-

```python
plt.scatter(x,y)
plt.title("Scatter Plot Example ")
plt.grid()
plt.show()
```

# Histogram :-

```python
data = [1,2,2,3,3,3,4,5]
plt.hist(data,bins = 5)
plt.title("Histogram chart Example")
plt.grid("True")
plt.show()
```
