```python
#import Essential  libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
import warnings
warnings.filterwarnings("ignore")
```

```python
#import dataset
df=pd.read_csv("student_info.csv")
df.head()
```

```python
df.shape
```

```python
#check for missing values.
df.isnull().sum()
```

```python
#Handle missing values.
df=df.fillna(df.mean())
df.isnull().sum()
```

```python
# visualization of data.
plt.scatter(x = df.study_hours,y=df.student_marks)
plt.title("Marks Prediction")
plt.xlabel("Study Hours")
plt.ylabel("Student Marks")
plt.show()
```

```python
#separate dataset into independent and dependent variables.
# X independent variable
X=df.drop("student_marks",axis=1)
X.head()
```

```python
#y dependent variable
y=df.drop("study_hours",axis=1)
y.head()
```

```python
#split dataset into training and testing dataset.
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#apply Machine learning algorithm.
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(X_train,y_train)
```

```python
#Test the Model.
res = model1.predict([[8.0]])[0][0].round(2)
print(f"expected result = {res}")
```

```python
#Test with input
sh = float(input("Enter study hours 4-12 : "))
if sh >=4 and sh<=12:
    res = model1.predict([[sh]])[0][0].round(2)
    print(f"Expected Result = {res}")
else:
    print("Invalid Study Hours")
    
```
