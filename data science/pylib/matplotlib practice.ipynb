```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
style.use("ggplot")
```

```python
#example application-1
x = [1,2,3]
y = [4,5,1]
plt.plot(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("my graph")
plt.show()
```

```python
#example Application-2
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'g',label='line 1',linewidth = 3)
plt.plot(x2,y2,'r',label = 'line 2',linewidth = 3)
plt.title('Epic Info')
plt.xlabel('X-Axis')
plt.ylabel('Y -Axis')
plt.grid(color = 'k')
plt.legend()
plt.show()
```

```python
# Example Application-3
students = ['Akram','Ariz','Ratnesh','Saif','Neha']
marks = [75,50,65,73,95]
plt.bar(students,marks,color= 'g')
plt.title('Marks Info')
plt.xlabel('student name')
plt.ylabel('Marks')
plt.grid(color='orange')
plt.show()
```

```python
 # Histogram Examples
marks=[45,55,60,65,70,72,75,80,85,90,95]
plt.hist(marks)
plt.title("Students Marks")
plt.xlabel("Marks")
plt.ylabel("Student")
plt.show()
```

```python
#Area plot Example.
months = ['jan','feb','march','April','May','Jun']
sales = [2000,2500,3000,2800,3500,4000]
plt.fill_between(months,sales,color = 'Orange')
plt.title("Sales of Months ")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(color = 'black')
plt.show()
```

```python
#LIne plot 
months = ['jan','feb','mar','apr','May','Jun']
sales = [120,150,170,160,180,200]
plt.plot(months,sales)
plt.title("Months of Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.show()
```
