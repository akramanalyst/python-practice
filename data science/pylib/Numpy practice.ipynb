```python
#import numpy library.
import numpy as np
lst1 = [1,2,3,4,5,6,7,8]
print(lst1)
```

```python
#one dimensional Array.
arr1= np.array(lst1)
print(arr1)
print(type(arr1))
```

```python
#two Dimensional Array.
lst2 = [1,2,3,4]
lst3 = [5,6,7,8]
arr2 = np.array([lst2,lst3])
print(arr2)
```

```python
#3 dimensional Array.
lst4 = [9,10,11,12]
arr3 = np.array([lst2,lst3,lst4])
print(arr3)
```

```python
#Reshape a 1-D array of size 6 into 2x3 matrix.
arr4 = arr1.reshape(2,4)
print(arr4)
```

```python
print("shape =",arr4.shape)
print("size = ",arr4.size)
print("dtype = ",arr4.dtype)
print("Dimension =",arr4.ndim)
```

```python
#zeros Array
arr5 = np.zeros(5)
print(arr5)
print(arr5.dtype)
```

```python
#ones Array.
arr6 = np.ones((4,4))
print(arr6)
```

```python
#full Array.
full_array = np.full((3,4),8)
print(full_array)
```

```python
#identity Array.
id_array = np.eye(4)
print(id_array)
```

```python
#empty Array.
print(np.empty(3))
```

```python
print(np.empty(2))
```

```python
print(np.empty(4))
```

```python
#Evenly spaced Array
print(np.arange(2,20,2))
```

```python
#specific number of equally(spaced) valuesbetween the range.
print(np.linspace(1,20,5))
```

```python
#random values Array.float
ran_array = np.random.rand(3,2)
print(ran_array)
```
