```python
import matplotlib.pyplot as plt
```

```python
#1 create a simple line plot for

```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
#2. add title ,X -labbel,y - label to a line plot
plt.plot(x,y)
plt.title("Line plot Example")
plt.xlabel("X Axis")
plt.ylabel(" y Axis")
plt.show()
```

```python
# 3 change the line style to (--) and add marker "o"
plt.plot(x,y,linestyle="--",marker = "o")
plt.title("line chart Example")
plt.xlabel("X axis")
plt.ylabel("y Axis")
plt.show()
```

```python
# 4 .plot two lines on same path .
y1 = [1,2,3,4,5]
y2 = [1,4,9,16,25]
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# 5 add a legend to distinguish multiple lines.
plt.plot(x,y1,label = "Line A")
plt.plot(x,y2,label = "Line B")
plt.legend()
plt.show()
```

```python
#6.save a plot as png file.
plt.plot(x,y1)
plt.plot(x,y2)
plt.savefig("my_plottsk.png")
plt.show()
```

```python
#7.create a bar chart for.
students = ["A","B","C","D"]
marks = [85,78,92,88]
plt.bar(students,marks)
plt.show()
```

```python
#8.create a scatter plot using :
x = [10,20,30,40,50]
y = [5,15,25,35,45]
plt.scatter(x,y)
plt.show()
```

```python
#9. change arker size and marker style in a scatter plot.
plt.scatter(x,y,linestyle="--",marker = "s",s=600)
plt.show()
```

```python
#10.add grid lines to a plot
plt.plot(x,y)
plt.grid(color = "green")
plt.show()
```

```python
#11.create a Histogram for a students marks.
marks = [45,56,78,88,90,67,76,85,92,73]
plt.hist(marks,bins=5)
plt.grid(color = 'k')
plt.show()
```
