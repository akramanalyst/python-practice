# common Array Creation functions:


```python
arr = np.zeros((2,3))   # This function creates a matrix 2x3 of all zeros.
print(arr)
```

```python
arr1 = np.ones((2,3))   # this function creates a matriz 2x3 of all ones.
print(arr1)
```

```python
arr2=np.eye(3)        # this function creates a identity matrix of 3x3 .
print(arr2)
```

```python
arr3=np.arange(1,10,2)  #this function generates numbers in given range.
print(arr3)
```

```python
arr4=np.linspace(1,5,5) #this function generates evenly spaced numbers.
print(arr4)    
```

```python
arr5=np.random.rand(3,3) # this function generates random numbers
print(arr5)
```

# Indexing and Slicing:

```python
arr = np.array([10,20,30,40,50])
print(arr) # [10 20 30 40 50]

```

```python
print(arr[0]) # 10
```

```python
print(arr[1:4]) # [20,30,40]
```

```python
mat = np.array([[1,2,3],[4,5,6]])
print(mat[1,1]) #5
```

```python

print(mat[:,1]) # all rows , column 1
```

# Array Operations.

```python
a = np.array([1,2,3])
b = np.array([4,5,6])
print(a+b)
```

```python
print(a*b)
```

```python
print(a**2)
```

# mathematical functions:-

```python
print(np.max(a))# display maximum elements of array a ---- 3
print(np.min(a))# display minimum elements os array a--- 1
print(np.mean(a))#display average of all elements of array a---2.0
print(np.sum(a))# display sum of all elements of array array a ---6
print(np.std(a))# display standard deviation.0.816496
```

# Linear Algebra Operations:


```python
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print(np.dot(A,B))# matrix multiplication
```

```python
print(np.linalg.det(A))# Determinant
```

```python
print(np.linalg.inv(A))# Inverse
```
